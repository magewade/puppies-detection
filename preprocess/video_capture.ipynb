```python
from pathlib import Path
import subprocess
```

```python
youtube_url = "https://www.youtube.com/watch?v=bYlEgU2tU5w"

# Получаем прямой URL потока
result = subprocess.run(["yt-dlp", "-g", youtube_url], capture_output=True, text=True)
stream_url = result.stdout.strip()

# Сохраняем видео
ffmpeg_cmd = [
    "ffmpeg",
    "-i",
    stream_url,
    "-t",
    "00:03:00",
    "-c",
    "copy",
    "puppies_train_toy_6.mp4",
]
subprocess.run(ffmpeg_cmd)
```

```python
input_video = "puppies_train_toy_6.mp4"
frame_dir = Path("frames")
frame_dir.mkdir(exist_ok=True)
```

```python
subprocess.run(
    ["ffmpeg", "-i", input_video, "-vf", "fps=0.5", f"{frame_dir}/frame_04_%04d.jpg"]
)
```
